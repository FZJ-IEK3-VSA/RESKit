```python
import numpy as np
import geokit as gk
```

```python
path = r"C:\Users\s.ryberg\PHD\repos\glaes\data\goodAreas.tif"

def makeGrid(res = 100):

    lons = np.arange(2100000, 7400001, res*1000, dtype=int)
    lats = np.arange(1200000, 5500001, res*1000, dtype=int)

    geoms = []
    for xi in range(len(lons)-1):
        for yi in range(len(lats)-1):
            ext = gk.Extent((lons[xi],lats[yi],lons[xi+1],lats[yi+1]), srs='europe_m')
            mat = ext.extractMatrix(path)
            mat = np.logical_or(mat==1, mat==2)

            if not mat.any(): continue
            
            geom = gk.geom.convertMask(mat, bounds=ext, flat=True)
            
            try:
                geom.Area()
                geoms.append(geom)
            except:
                pass

    if len(geoms)>0: gk.vector.createVector(geoms, output="grid_%dkm.shp"%res, overwrite=True)
```

```python
makeGrid(10)
makeGrid(25)
makeGrid(50)
makeGrid(100)
```
